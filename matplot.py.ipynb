```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
import numpy as np
```

```python
# obiekt klasy Figure
fig = plt.figure()
# obiekt klasy Axis, bez niego canvas nie ma osi, więc się nie pokazuje
ax = plt.axes()
```

```python
fig_1 = plt.figure()
ax_1 = plt.axes()
# określamy skalę x, linspace daje 1000 równo rozmieszczonych punktów w przedziale 0-10
x = np.linspace(0, 10, 1000)
# na osi x możemy plotować różne funkcje
ax_1.plot(x, np.sin(x))
ax_1.plot(x, np.cos(x))
```
