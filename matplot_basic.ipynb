```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
import numpy as np
```

## Obiekty Figure i Axis

```python
# obiekt klasy Figure
fig = plt.figure()
# obiekt klasy Axis, bez niego canvas nie ma osi, więc się nie pokazuje
ax = plt.axes()
```

## Punkty dla x oraz ustawienia plot()

```python
fig_1 = plt.figure()
ax_1 = plt.axes()
# określamy skalę x, linspace daje 1000 równo rozmieszczonych punktów w przedziale 0-10
x = np.linspace(0, 10, 1000)
# na osi x możemy plotować różne funkcje
# można określić kolor i kreskowanie "-", "--", "-.", ":"
ax_1.plot(x, np.sin(x), color='orange', linestyle='--')
ax_1.plot(x, np.cos(x))
```

## Ustawiamy skale osi

```python
fig_2 = plt.figure()
ax_2 = plt.axes()
x = np.linspace(0, 10, 1000)
ax_2.plot(x, np.sin(x))
# możemy ustawić dowolną skalę osi, a nawet ją odwrócić (xlim)
ax_2.set_xlim(11, -1)
ax_2.set_ylim(-1.5, 1.5)
# poniższe dostosuje skalę do wykresu
# ax_2.axis('tight')
# poniższe ustawi wykres y=x
# ax_2.axis('equal')
```

## Opisywanie wykresu i legenda

```python
fig_3 = plt.figure()
ax_3 = plt.axes()
x = np.linspace(0, 10, 1000)
# label idzie do legendy
ax_3.plot(x, np.sin(x), label='sin(x)')
ax_3.axis('equal')
ax_3.set_title('A Sine curve')
ax_3.set_xlabel('Variable x')
ax_3.set_ylabel('Sin(x)')
ax_3.legend()
# wszystko powyższe można zrobić za jednym razem
# ax_3.set(xlim=(0,10), ylim=(-2, 2), xlabel='x', ylabel='sin(x)', title='A Sine curve')
```

## Scatter plots

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

# fig_4 = plt.figure()
# ax_4 = plt.axes()
# ax_4.plot(x, y, 'o', color='yellow')

# albo po prostu

plt.plot(x, y, 'o', color='orange')

# rodzaje markerów: o . x + < > s d v ^
# można je też połączyć z linią, np. '-o' da linię i kropki
# są też inne opcje modyfikujące markery
# markersize=15
# linewidth=4
# markerfacecolor='white'
# markeredgecolor='gray'
# markeredgewidth=2
```

# plt.scatter()

plt.plot() jest wydajniejszy, bo każdy punkt wizualizuje tak samo. plt.scatter() może różnicować poszczególne punkty, więc jest wolniejszy

```python
# obiekt generatora
rng = np.random.default_rng(0)
# inicjalizacja generatora z rozkładem normalnym
x = rng.normal(size=100)
y = rng.normal(size=100)
# randomowe kolory?
colors = rng.random(100)
# rozmiar w pikselach
sizes = 1000 * rng.random(100)

# alpha czyli przeźroczystość
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3)
plt.colorbar()
```

## Error bars

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

# wykres scatter z errorami plus opcje wyświetlenia
plt.errorbar(x, y, yerr=dy, fmt='o', color='orange', ecolor='green', elinewidth=2, capsize=3)
```

## Histograms


```python
# obiekt generatora
rng = np.random.default_rng(1701)
# generujemy dane o rozkładzie normalnym
data = rng.normal(size=1000)
# histogram z opcjami
plt.hist(data, bins=30, alpha=0.8, density=True, color='steelblue', edgecolor='yellow')
```

```python
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 500)
x3 = rng.normal(3, 2, 1500)

# density True oznacza, że całka=1, czyli pokazuje gęstość prawdopodobieństwa
kwargs = dict(histtype='barstacked', alpha=0.3, density=True, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

```python
# liczymy wartości histogramu
counts, bin_edges = np.histogram(data, bins=10)
counts
```

## Histogram dwuwymiarowy

```python
# dane dla x i y wg normalnego rozkładu Gaussa dla dwóch zmiennych
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = rng.multivariate_normal(mean, cov, 10000).T
# histogram 2d
plt.hist2d(x, y, bins=30)
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
# liczymy wartości dla histogramu2d
counts, xedges, yedges = np.histogram2d(x, y, bins=30)
counts
```

```python
# można też zrobić histogram heksagonalny
plt.hexbin(x, y, gridsize=25)
cb = plt.colorbar(label='count in bin')
```

## Kernel Density Estimation

```python
from scipy.stats import gaussian_kde

# fit array of size (num of dimensions, number of samples
data = np.vstack([x, y])
kde = gaussian_kde(data)

# evaluate on a grid
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# plot the results as image
plt.imshow(Z.reshape(Xgrid.shape), origin='lower', aspect='auto', extent=[-3.5, 3.5, -6, 6])
cb = plt.colorbar()
cb.set_label('density')
```
