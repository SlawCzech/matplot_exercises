# plt.scatter()

plt.plot() jest wydajniejszy, bo każdy punkt wizualizuje tak samo. plt.scatter() może różnicować poszczególne punkty, więc jest wolniejszy

```python
# obiekt generatora
rng = np.random.default_rng(0)
# inicjalizacja generatora z rozkładem normalnym
x = rng.normal(size=100)
y = rng.normal(size=100)
# randomowe kolory?
colors = rng.random(100)
# rozmiar w pikselach
sizes = 1000 * rng.random(100)

# alpha czyli przeźroczystość
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3)
plt.colorbar()
```

# Error bars

# Histograms


```python
# obiekt generatora
rng = np.random.default_rng(1701)
# generujemy dane o rozkładzie normalnym
data = rng.normal(size=1000)
# histogram z opcjami
plt.hist(data, bins=30, alpha=0.8, density=True, color='steelblue', edgecolor='yellow')
```

```python
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 500)
x3 = rng.normal(3, 2, 1500)

# density True oznacza, że całka=1, czyli pokazuje gęstość prawdopodobieństwa
kwargs = dict(histtype='barstacked', alpha=0.3, density=True, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

```python
# liczymy wartości histogramu
counts, bin_edges = np.histogram(data, bins=10)
counts
```

# Histogram dwuwymiarowy

```python
# dane dla x i y wg normalnego rozkładu Gaussa dla dwóch zmiennych
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = rng.multivariate_normal(mean, cov, 10000).T
# histogram 2d
plt.hist2d(x, y, bins=30)
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
# liczymy wartości dla histogramu2d
counts, xedges, yedges = np.histogram2d(x, y, bins=30)
counts
```

```python
# można też zrobić histogram heksagonalny
plt.hexbin(x, y, gridsize=25)
cb = plt.colorbar(label='count in bin')
```

# Kernel Density Estimation

```python
from scipy.stats import gaussian_kde

# fit array of size (num of dimensions, number of samples
data = np.vstack([x, y])
kde = gaussian_kde(data)

# evaluate on a grid
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# plot the results as image
plt.imshow(Z.reshape(Xgrid.shape), origin='lower', aspect='auto', extent=[-3.5, 3.5, -6, 6])
cb = plt.colorbar()
cb.set_label('density')
```

# Text and annotations

```python
from datetime import datetime

births = pd.read_csv('data/births.csv')

quartiles = np.percentile(births['births'], [25, 50, 75])

mu, sig = quartiles[1], 0.74 * (quartiles[2] - quartiles[0])
births = births.query('(births > @mu -5 * @sig) & (births < @mu + 5 * @sig)')

births['day'] = births['day'].astype(int)

births.index = pd.to_datetime(10000 * births.year + 100 * births.month + births.day, format='%Y%m%d')
births_by_date = births.pivot_table('births', [births.index.month, births.index.day])
births_by_date.index = [datetime(2012, month, day) for (month, day) in births_by_date.index]
```

```python
import matplotlib as mpl
fig, ax = plt.subplots(figsize=(14, 4))
births_by_date.plot(ax=ax)

# słownik do stylizacji napisu
style = dict(size=10, color='gray')

# lokalizacja napisów
ax.text('2012-1-1', 3950, "New Year's Day", **style)
ax.text('2012-7-4', 4250, "Independence Day", ha='center', **style)
ax.text('2012-9-4', 4850, "Labor Day", ha='center', **style)
ax.text('2012-10-31', 4600, "Halloween", ha='right', **style)
ax.text('2012-11-25', 4400, "Thanksgiving", ha='center', **style)
ax.text('2012-12-25', 3850, "Christmas", ha='right', **style)

# nazywanie osi y i wykresu
ax.set(title='USA births by day of year (1969-1988', ylabel='average daily births')

# formatowanie osi x
ax.xaxis.set_major_locator(mpl.dates.MonthLocator())
ax.xaxis.set_minor_locator(mpl.dates.MonthLocator(bymonthday=15))
ax.xaxis.set_major_formatter(plt.NullFormatter())
ax.xaxis.set_minor_formatter(mpl.dates.DateFormatter('%h'))
```
